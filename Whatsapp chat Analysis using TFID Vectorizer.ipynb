```python
# creating a func to see the results of sentiment analysis
def sentimentalAnalysis(dataset,columnname): # dataset and columname as input
    # import libs
    import matplotlib.pyplot as plt
    import seaborn as sns
    import nltk
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    # initialize var and assign lib 
    sid = SentimentIntensityAnalyzer()
    # remove null
    dataset.dropna(inplace =True)
    # senti = polarity_scores(df['Chat'][2]) # fo single col
    # to get a scores for all column by using lambda func
    # adding columns to the dataset: scores, compound, negative, positive, neutral
    dataset['scores']= dataset[columnname].apply(lambda commentText: sid.polarity_scores(commentText)) # columnname = chat, commentText = message in that chat
    # the above code will find the scores for all the col's
    # for single row {'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0} - this will be the result, so based on this need to add col
    dataset['compound']=dataset['scores'].apply(lambda score_dict: score_dict['compound']) # compound col from scores 
    dataset['Negative']=dataset['scores'].apply(lambda score_dict: score_dict['neg'])
    dataset['Positive']=dataset['scores'].apply(lambda score_dict: score_dict['pos'])
    dataset['Neutral']=dataset['scores'].apply(lambda score_dict: score_dict['neu'])
    
#   Creating final positive or negative using compound score
    dataset['comp_score'] = dataset['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
    plt.clf()

#   Checking how many positive and negative
    posneg=pd.DataFrame(dataset['comp_score'].value_counts())
    return posneg,dataset
# sentimentalAnalysis(dataset,columnname ='Chat')   
```

```python
pos,data_senti = sentimentalAnalysis(dataset,columnname ='Chat')   
pos,data_senti
```

# Topic Modelling

```python
# !pip install scikit-learn
```

```python
# importing TFID vector for preprocessing 
# input for topic modelling should be vector
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# tfid 
tfidf = TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm = tfidf.fit_transform(df['Chat']) # fit and transform  --> chat col will get fit and the prediction for chat ccol will also be excuted
dtm
```

```python
# NLP algorithm NMF and LDA
from sklearn.decomposition import NMF # non matrix factorization in decomposition
nmf_model = NMF(n_components = 5, random_state=42) #n_components shows the no of topics, random_state to get an same o/p
nmf_model.fit(dtm) # model fit
```

```python
# to check the results of topics
for index, topic in enumerate(nmf_model.components_): # enumerate will create the index number based on the need and the condition
    results = ([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]]) # get_feature_names_out will give the topic names
    print(results) # -10 will show the number of words

```

```python
# to predict
topic_results = nmf_model.transform(dtm)
# to add the topic in df
df['Topic']= topic_results.argmax(axis =1)
```

```python
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
import matplotlib.pyplot as plt
nltk.download('stopwords')
```

```python
# creating custom words empty list
custom_words =[]
# creating stopwords list 
stoplist = stopwords.words('english')
# stoplist 
# extend the stopwords by adding extra 
stoplist.extend(['omitted', 'voice','missed','call','video','deleted','media','message'])
# stoplist


#  to create the word cloud
df.index = range(df.shape[0]) # to get the shape of df

for i in range(1,len(df)):  # for i in 1 to 50
    custom_words.append(df['Chat'][i]) # custom words will get append with df[chat]
    pic =" ".join(custom_words) # custom words will get add
    wordcloud = WordCloud(width = 800, height = 800, 
                         background_color = 'white',
                         stopwords = stoplist,
                         min_font_size = 10).generate(pic)

plt.figure(figsize = (9, 7), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.savefig('wordcloud.PNG')
plt.show() 
print("Successfully created")
  
```

# Topic modelling using LDA (Latent Dirichlet Allocation)

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model = LatentDirichletAllocation(n_components = 5, random_state = 42)
lda_model .fit(dtm)
```

```python
# to check the results of topics
for index, topic in enumerate(lda_model.components_): # enumerate will create the index number based on the need and the condition
    words = ([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]]) # get_feature_names_out will give the topic names
    print(f"Topic {index + 1}: {words}") # -10 will show the number of words
```

```python
# to predict
topic_results = lda_model.transform(dtm)
# to add the topic in df
df['Topic']= topic_results.argmax(axis =1)
```
