```python
# import lib
import pandas as pd
#load dataset
dataset = pd.read_csv("mabel.txt",header=None,on_bad_lines='skip',encoding ='utf8')
```

# Pre - Processing

```python
df =dataset
df = df.drop(0)
df.columns= ['Date','Chats']
Time = df["Chats"].str.split("-",n = 1,expand =True) # n=1 is like first - 
df["Time"] = Time[0] # in time var it is stored in 0 column
Message = Time[1].str.split(":",n=1,expand =True) # message var = timevar[1] 1st column contains names and mesages
df['Name'] = Message[0]
df['Chat'] = Message[1]
df =df.drop('Chats',axis =1)
df
```

```python
df.info()
```

# Sentimental analysis

```python
# !pip install nltk
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df['Chat'][5]
```

```python
sid = SentimentIntensityAnalyzer()
senti = sid.polarity_scores(df['Chat'][5])
senti
```

```python
def sentimentalAnalysis(df, columnname):
    sid = SentimentIntensityAnalyzer()
    df.dropna(inplace =True)
    df['scores']= df[columnname].apply(lambda commentText: sid.polarity_scores(commentText)) # columnname = chat, commentText = message in that chat
    df['compound']=df['scores'].apply(lambda score_dict: score_dict['compound'])
    df['Negative']=df['scores'].apply(lambda score_dict: score_dict['neg'])
    df['Positive']=df['scores'].apply(lambda score_dict: score_dict['pos'])
    df['Neutral']=df['scores'].apply(lambda score_dict: score_dict['neu'])
    
    df['comp_score'] = df['compound'].apply(lambda c:'pos' if c>= 0 else 'neg')    
    posneg = pd.DataFrame(df['comp_score'].value_counts())
    return posneg,df
    
```

```python
pos,data_senti = sentimentalAnalysis(df,columnname ='Chat')
pos,data_senti
```

# Hashing Vector

```python
# importing TFID vector for preprocessing 
# input for topic modelling should be vector
from sklearn.feature_extraction.text import HashingVectorizer
hash_vect = HashingVectorizer(stop_words='english', n_features=10000, alternate_sign=False, norm=None)  # Adjust n_features as needed
htm = hash_vect.fit_transform(df['Chat'])
htm
```

# Topic Modelling using NMF

```python
# NLP algorithm NMF and LDA
from sklearn.decomposition import NMF # non matrix factorization in decomposition
nmf_model = NMF(n_components = 5, random_state=42) #n_components shows the no of topics, random_state to get an same o/p
nmf_model.fit(htm) # model fit
```

```python
# to check the results of topics
# Display the top words for each topic
for index, topic in enumerate(nmf_model.components_):
    top_indices = topic.argsort()[-10:]
    top_words = [f"Feature {i}" for i in top_indices]  # HashingVectorizer doesn't provide feature names
    print(f"Topic {index + 1}: {top_words}")
```

```python
# to predict
topic_results = nmf_model.transform(htm)
# to add the topic in df
df['Topic']= topic_results.argmax(axis =1)
```

```python
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
import matplotlib.pyplot as plt
nltk.download('stopwords')
```
